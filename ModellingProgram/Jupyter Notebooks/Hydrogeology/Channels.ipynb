```python
def descent(arr, Pos, carryOver):
    '''
    This function is a more concise version of the previous one that also incorporates
    diagonal flow somewhat. Really, it's just easier to read
    
    - carryOver represents how close the previous values were to moving over 1 unit
        - normalized to the largest value
    
    - It may be somewhat slower, but I do not expect this function to be run
        very often, so I have not taken much time in making it faster
    '''
    width = arr.shape[1] - 1
    depth = arr.shape[0] - 1
    
    currentHeight = arr[Pos]

    atBottom = False

#---------------------------------------------------------------------------------
    # Check for corner
    if Pos in [(0,0), (0,width), (depth,0), (depth,width)]:
        # Top Left corner
        if Pos == (0,0):
            Down = (arr[Pos[0] + 1, Pos[1]], [Pos[0] + 1, Pos[1]], (1,0))
            Right = (arr[Pos[0], Pos[1] + 1], [Pos[0], Pos[1] + 1], (0,1))
            
            vals = [Down, Right]
            
            if vals[0][0] > vals[1][0]:
                vals[0], vals[1] = vals[1], vals[0]
                
            if vals[0][0] > currentHeight:
                atBottom = True
                return Pos, carryOver, atBottom
            
            newPosition = vals[0][1]
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow
                
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

                #print("executed carryover code")

            return tuple(newPosition), carryOver, atBottom
            
            
            
#-----------------------------------  
        # Top Right corner
        elif Pos == (0,width):
            Down = (arr[Pos[0] + 1, Pos[1]], [Pos[0] + 1, Pos[1]], (1,0))
            Left = (arr[Pos[0], Pos[1] - 1], [Pos[0], Pos[1] - 1], (0,-1))
            
            vals = [Down, Left]
            
            if vals[0][0] > vals[1][0]:
                vals[0], vals[1] = vals[1], vals[0]
                
            if vals[0][0] > currentHeight:
                atBottom = True
                return Pos, carryOver, atBottom
            
            newPosition = vals[0][1]
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow
                
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

                #print("executed carryover code")

            return tuple(newPosition), carryOver, atBottom
            
        
#-----------------------------------
        # Bottom left corner
        elif Pos == (depth, 0):
            Up = (arr[Pos[0] - 1, Pos[1]], [Pos[0] - 1, Pos[1]], (-1,0))
            Right = (arr[Pos[0], Pos[1] + 1], [Pos[0], Pos[1] + 1], (0,1))
            
            vals = [Up, Right]
            
            if vals[0][0] > vals[1][0]:
                vals[0], vals[1] = vals[1], vals[0]
                
            if vals[0][0] > currentHeight:
                atBottom = True
                return Pos, carryOver, atBottom
            
            newPosition = vals[0][1]
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow
                
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

                #print("executed carryover code")

            return tuple(newPosition), carryOver, atBottom

#-----------------------------------
        # Bottom right corner
        else:
            Up = (arr[Pos[0] - 1, Pos[1]], [Pos[0] - 1, Pos[1]], (-1,0))
            Left = (arr[Pos[0], Pos[1] - 1], [Pos[0], Pos[1] - 1], (0,-1))
            
            vals = [Up, Left]
            
            if vals[0][0] > vals[1][0]:
                vals[0], vals[1] = vals[1], vals[0]
                
            if vals[0][0] > currentHeight:
                atBottom = True
                return Pos, carryOver, atBottom
            
            newPosition = vals[0][1]
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow
                
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

                #print("executed carryover code")

            return tuple(newPosition), carryOver, atBottom

#---------------------------------------------------------------------------------
    # Check for top or bottom edge
    elif Pos[0] == 0 or Pos[0] == depth:
        # Do top edge
        if Pos[0] == 0:
            Left = (arr[Pos[0], Pos[1] - 1], [Pos[0], Pos[1] - 1], (0,-1))
            Right = (arr[Pos[0], Pos[1] + 1], [Pos[0], Pos[1] + 1], (0,1))
            Down = (arr[Pos[0] + 1, Pos[1]], [Pos[0] + 1, Pos[1]], (1,0))
            
            vals = [Left, Right, Down]
            
            # Ordering the values in increasing order
            while not (vals[0][0] <= vals[1][0] <= vals[2][0]):
                for i in range(2):
                    if vals[i][0] > vals[i + 1][0]:
                        vals[i],vals[i+1] = vals[i+1],vals[i]
            
            # Check to see if at bottom
            if vals[0][0] > currentHeight:
                #print("At the bottom")
                return Pos, carryOver, True    # True refers to being at a local minimum
            
            # Determine the new position
            newPosition = vals[0][1]
            
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow


            # Check to see if carryOver > 1. If so, flow 1 step in the direction of vals[1]
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

                #print("executed carryover code")

            return tuple(newPosition), carryOver, False
        
#---------------------------------------------------------------------------------
        # Do bottom edge
        else:
            Up = (arr[Pos[0] - 1, Pos[1]], [Pos[0] - 1, Pos[1]], (-1,0))
            Left = (arr[Pos[0], Pos[1] - 1], [Pos[0], Pos[1] - 1], (0,-1))
            Right = (arr[Pos[0], Pos[1] + 1], [Pos[0], Pos[1] + 1], (0,1))
            
            vals = [Up, Left, Right]
            
            # Ordering the values in increasing order
            while not (vals[0][0] <= vals[1][0] <= vals[2][0]):
                for i in range(2):
                    if vals[i][0] > vals[i + 1][0]:
                        vals[i],vals[i+1] = vals[i+1],vals[i]
            
            # Check to see if at bottom
            if vals[0][0] > currentHeight:
                #print("At the bottom")
                return Pos, carryOver, True    # True refers to being at a local minimum
            
            # Determine the new position
            newPosition = vals[0][1]
            
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow


            # Check to see if carryOver > 1. If so, flow 1 step in the direction of vals[1]
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

                #print("executed carryover code")

            return tuple(newPosition), carryOver, False
        
#---------------------------------------------------------------------------------
    # Check for left or right edge
    elif Pos[1] == 0 or Pos[1] == width:
#---------------------------------------------------------------------------------
        # Check left
        if Pos[1] == 0:
            Up = (arr[Pos[0] - 1, Pos[1]], [Pos[0] - 1, Pos[1]], (-1,0))
            Down = (arr[Pos[0] + 1, Pos[1]], [Pos[0] + 1, Pos[1]], (1,0))
            Right = (arr[Pos[0], Pos[1] + 1], [Pos[0], Pos[1] + 1], (0,1))
            
            vals = [Up, Down, Right]
            
            # Ordering the values in increasing order
            while not (vals[0][0] <= vals[1][0] <= vals[2][0]):
                for i in range(2):
                    if vals[i][0] > vals[i + 1][0]:
                        vals[i],vals[i+1] = vals[i+1],vals[i]
            
            # Check to see if at bottom
            if vals[0][0] > currentHeight:
                #print("At the bottom")
                return Pos, carryOver, True    # True refers to being at a local minimum
            
            # Determine the new position
            newPosition = vals[0][1]
            
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow


            # Check to see if carryOver > 1. If so, flow 1 step in the direction of vals[1]
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

                #print("executed carryover code")

            return tuple(newPosition), carryOver, False
        
#---------------------------------------------------------------------------------
        # Check right
        else:
            Up = (arr[Pos[0] - 1, Pos[1]], [Pos[0] - 1, Pos[1]], (-1,0))
            Down = (arr[Pos[0] + 1, Pos[1]], [Pos[0] + 1, Pos[1]], (1,0))
            Left = (arr[Pos[0], Pos[1] - 1], [Pos[0], Pos[1] - 1], (0,-1))
            
            vals = [Up, Left, Left]
            
            # Ordering the values in increasing order
            while not (vals[0][0] <= vals[1][0] <= vals[2][0]):
                for i in range(2):
                    if vals[i][0] > vals[i + 1][0]:
                        vals[i],vals[i+1] = vals[i+1],vals[i]
            
            # Check to see if at bottom
            if vals[0][0] > currentHeight:
                #print("At the bottom")
                return Pos, carryOver, True    # True refers to being at a local minimum
            
            # Determine the new position
            newPosition = vals[0][1]
            
            standard = currentHeight - vals[0][0]
            if standard == 0:
                standard = .00001
            another = currentHeight - vals[1][0]
            additionalFlow = another/standard
            if vals[1][0] < currentHeight:
                carryOver += additionalFlow


            # Check to see if carryOver > 1. If so, flow 1 step in the direction of vals[1]
            if carryOver >= 1.0:
                carryOver -= 1.0
                additionalDirection = vals[1][2]

                newPosition[0] += additionalDirection[0]
                newPosition[1] += additionalDirection[1]

               #print("executed carryover code")

            return tuple(newPosition), carryOver, False
            
            
#---------------------------------------------------------------------------------
        # Do any of the interior
    else:
        #print("doing else")
        Up = (arr[Pos[0] - 1, Pos[1]], [Pos[0] - 1, Pos[1]], (-1,0))
        Down = (arr[Pos[0] + 1, Pos[1]], [Pos[0] + 1, Pos[1]], (1,0))
        Left = (arr[Pos[0], Pos[1] - 1], [Pos[0], Pos[1] - 1], (0,-1))
        Right = (arr[Pos[0], Pos[1] + 1], [Pos[0], Pos[1] + 1], (0,1))
        
        vals = [Up, Down, Left, Right]
        
        # Ordering the values in increasing order
        while not (vals[0][0] <= vals[1][0] <= vals[2][0] <= vals[3][0]):
            for i in range(3):
                if vals[i][0] > vals[i + 1][0]:
                    vals[i],vals[i+1] = vals[i+1],vals[i]
                    
        # Checking if at bottom
        if vals[0][0] > currentHeight:
            #print("At the bottom")
            return Pos, carryOver, True    # True refers to being at a local minimum
        
        # Determine the new position
        newPosition = vals[0][1]
        
        # If the value made it to this, the first value in the list is the smallest and therefore the reference
        # I use that value as the standard
        # At the moment, I only use the second value in the list to determine a 2D flow direction
        standard = currentHeight - vals[0][0]
        #print("Standard: ", standard)
        if standard == 0:
            standard = .00001
        another = currentHeight - vals[1][0]
        #print("Another: ", another)
        additionalFlow = another/standard
        if vals[1][0] < currentHeight:
            carryOver += additionalFlow
        
        
        # Check to see if carryOver > 1. If so, flow 1 step in the direction of vals[1]
        if carryOver >= 1.0:
            carryOver -= 1.0
            additionalDirection = vals[1][2]
            
            newPosition[0] += additionalDirection[0]
            newPosition[1] += additionalDirection[1]
            
            #print("executed carryover code")
            
        return tuple(newPosition), carryOver, False
```
